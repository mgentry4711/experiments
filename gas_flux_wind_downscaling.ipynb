```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def u_squared(U, R, ri):
    ex = -5.0/3.0 * (1.0/ri - 1.0/R)
    u2 = (R**3.0/ri**3.0) * U**2.0 * np.exp(ex)
    return np.sqrt(u2)
    
```

```python
R = 100e3 # 100 km grid cell
ri = np.linspace(50e3, 200e3, 1000)
U = 10.0 # m s^-1

u2 = u_squared(U, R, ri)
```

```python
for U in np.linspace(0, 30, 10):
    u2 = u_squared(U, R, ri)
    plt.plot(ri, u2)
    plt.plot(R, U, 'o')
```
