```python
import numpy as np
```

```python
### np.genfromtext is useful for opening text files
path = '/Users/matthewgentry/Downloads/ATOC5235_hw2.txt'
ff = np.genfromtxt(path, delimiter='\t')

```

```python
### the csv module is good for text files or .csv files
### you can save sheets as .csv files in Excel
import csv
with open(path, 'r') as f:
    c = csv.reader(f, delimiter='\t') # this means the file is tab-separated
    data = []
    for i, line in enumerate(c):
        data.append(line)
```

```python
### This bit labels the data clearly in a thing called a dictionary
skip_header = 2 # skip this many rows before getting to numeric values
labeled_data = {}
for name in data[0]:
    labeled_data[name] = np.array(data).T[0][skip_header:].astype(float)
    # this last line is the tricky one that chooses the right section of the data
    # and labels it with "name" for each name in the first row of "data"
    # python uses zero [0] to denote the first entry in a list or array 
```

```python
import matplotlib.pyplot as plt # this is for plotting 
### With clear labels, we can now use "if" to decide what to plot
### without needing to know which row or column it started in.
plt.plot(labeled_data['lambda'], labeled_data['absorption'])
# you can copy and past this line as many times as you need
```

```python
# or do something like:

for entry in labeled_data:
    print('entry =', entry) # just so you can see it
    fig = plt.figure() # makes a new figure to start editing
    plt.plot(labeled_data['lambda'], labeled_data[entry])
    plt.show()
    fig.savefig('./where_you_want_to_save_this_plot.png')

# to plot all the columns in a file against the variable "lambda"
```
